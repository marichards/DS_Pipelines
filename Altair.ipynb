```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import altair as alt
```

```python
# Load the Iris dataset
iris = alt.load_dataset('iris')
print(iris.shape)
print(iris.head())
```

**Note: to make plots with Altair, use the "Chart" object**

```python
iris_chart = alt.Chart(iris)
type(iris_chart)
```

## Histograms

This is from a recipe on the Altair documentation:

```python
iris_chart.mark_bar().encode(
    x = alt.X('petalWidth', 
              bin = alt.Bin(maxbins = 3),
             ),
    y = 'count(*):Q'
)
```

*Some notes on this:*

- The maxbins only sets the max number; the chart will take whatever number fits below that, I think
- The "Q" notation is for "quantitative"

```python
iris_chart.mark_bar().encode(
    x = alt.X('petalWidth', 
              bin = alt.Bin(maxbins = 3),
             ),
    y = 'count(*):Q',
    color = 'species'
)
```

The above example shows a histogram colored by value

## Scatter plots

Here's a simple histogram, colored by species:

```python
iris_chart.mark_circle().encode(
    x = 'petalWidth',
    y = 'sepalLength',
    color = 'species'
)
```
